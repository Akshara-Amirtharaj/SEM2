```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
data=pd.read_csv(r"C:\Users\aksha\Desktop\AKSHARA\SEM 2\DATA SCIENCE\Proper\Linear regression\cars.csv")
```

```python
df1=pd.DataFrame(data)
df1
```

```python
encode=LabelEncoder()
```

```python
df1['condition']=encode.fit_transform(df1['condition'])
df1['model']=encode.fit_transform(df1['model'])
df1['make']=encode.fit_transform(df1['make'])
df1['fuel_type']=encode.fit_transform(df1['fuel_type'])
df1['color']=encode.fit_transform(df1['color'])
df1['transmission']=encode.fit_transform(df1['transmission'])
df1['drive_unit']=encode.fit_transform(df1['drive_unit'])
df1['segment']=encode.fit_transform(df1['segment'])
df1
```

```python
df1['volume(cm3)'].fillna(df1['volume(cm3)'].mean(),inplace=True)
```

```python
Q1 = df1.quantile(0.25)
Q3 = df1.quantile(0.75)
IQR = Q3 - Q1
df = df1[~((df1< (Q1 - 1.5 * IQR)) |(df1 > (Q3 + 1.5 * IQR))).any(axis=1)]
```

```python
x=df.drop('priceUSD',axis=1)
y=df['priceUSD']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25)
```

```python
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
print(model.score(x_train,y_train)*100)
print(model.score(x_test,y_test)*100)
```
